```python
import os

from dotenv import load_dotenv
```

```python
load_dotenv('../.env')
API_KEY = os.getenv('API_KEY')
```

```python
import osmnx as ox
```

```python
G = ox.graph_from_place('Zurich, Switzerland', network_type='walk')
```

```python
G = ox.add_node_elevations_google(G, api_key=API_KEY)
```

```python
fig, ax = ox.plot_graph(G)
```

```python
for node, data in G.nodes(data=True):
    print(data)
```

```python
ox.save_graphml(G, './data/zurich_walk.graphml')
```
